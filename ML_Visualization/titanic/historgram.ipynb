```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
import numpy as np
```

```python
df=sns.load_dataset('mpg')
sns.set(font=('D2Coding ligature'))
```

```python
df.rename(columns={'mpg':'연비','cylinders':'실린더','displacement':'배기량','horsepower':'마력','weight':'중량','acceleration':'가속','model_year':'년식','origin':'생산지'})
```

```python
df.info()
```

```python
df.columns = ['연비','실린더','배기량','마력','중량', '가속','년식','제조국','이름']
df
```

```python
df.info()
```

```python
df.hist(figsize=(10,10), bins=10)
```

```python
# 연비에 대한 히스토그램. kde 선형 그래프 함께표시 T or F
sns.displot(data=df,x='연비',kde=True)
```

```python
# shade 그래프 내부 음영처리
sns.kdeplot(data=df,x='연비', shade=True)
```

```python
# 연비 값에 대해 skew(왜도,skewness), kutr(첨도,kurtosis)값 구하기
# 하나의 칼럼 객체 -> 시리즈(Series)
df['연비'].agg(['skew','kurt'])
```

```python
# 데이터프레임의 무작위 데이터 100개 셈플 채취
df_sample=df.sample(100,random_state=42)
# 제조국별 데이터 항목별 비교차트
sns.pairplot(data=df_sample,hue='제조국')
```
