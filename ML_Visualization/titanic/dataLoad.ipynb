```python
import pandas as pd
import seaborn as sns

import matplotlib as plt


```

```python
# seaborn 폰트 깨질때 폰트 변경
sns.set(font='D2Coding ligature')
```

```python
df=pd.read_csv('./dataframes/train.csv',encoding='cp949')
```

```python
df.rename(columns={'PassengerId':'승객번호','Survived' : '생존여부', 'Pclass' : '좌석등급', 'Name':'이름', 'Sex':'성별','Age':'나이', 'SibSp' : '혈족', 'Parch':'직계','Ticket':'승선권', 'Fare':'요금', 'Cabin':'객실번호','Embarked':'승선항'}, inplace=True)
```

```python
sns.countplot(data=df,x='좌석등급')
```

```python
sns.countplot(data=df,x='생존여부')
```

```python
sns.countplot(data=df,x='성별')
```

```python
sns.countplot(data='df', y='나이')
```

```python
df.hist(figsize=(10,10), bins=10)
```

```python
# 비대칭도 (왜도)
# 양수(+) <-- (평균=중앙값) -->음수(-)
df.skew()
```

```python
# 첨도
# 데이터가 어느 곳에 집중적으로 몰려있는가를 측정
# k<3<k : k값이 3에 가까울 수록 정규분포에 가깝다.
df.kurt(axis='생존여부')
```

```python
sns.displot(data=df,x='나이',kde=True)
```
